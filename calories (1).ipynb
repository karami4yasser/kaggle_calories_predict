```python
text=df["Activity, Exercise or Sport (1 hour)"].values
text[0:10]
```

```python
letters=[]
letters_uniques=[]
for ligne in text:
  lettre=ligne.split(",")
  for i in lettre:
      letters.append(i)
      

```

```python
letters_uniques = list(dict.fromkeys(letters))
```

```python
letters_uniques.count(letters[100])
```

```python
print(len(letters_uniques))
print(len(letters))
```

```python
letters_uniques_with_fre=[]
for i in range(len(letters_uniques)):
  letters_uniques_with_fre.append((letters_uniques[i],letters.count(letters_uniques[i])))
```

```python
def takeSecond(elem):
    return elem[1]
letters_uniques_with_fre.sort(key=takeSecond,reverse=True)
```

```python
letters_uniques_with_fre
```

```python
df.head(2)
```

```python
for i in range(len(df)):
    for j in range(len(letters_uniques_with_fre)):
        if letters_uniques_with_fre[j][0] in df.loc[i,'Activity, Exercise or Sport (1 hour)']:
            df.loc[i,letters_uniques_with_fre[j][0]]=1
        else:
          df.loc[i,letters_uniques_with_fre[j][0]]=0
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.feature_selection import mutual_info_regression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
from hyperopt import hp,fmin,tpe,STATUS_OK,Trials
from sklearn.metrics import mean_squared_error,mean_absolute_error
import numpy as np
```

```python
rf = RandomForestRegressor()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x[x.columns[:70]], y, test_size=0.3, random_state=0)
```

```python
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 12000, num = 120)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(5, 3000, num = 60)]
min_samples_split = [2, 5, 10, 15, 100]
min_samples_leaf = [1, 2, 5, 10]
```

```python
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
y_pred = rf_random.predict(X_test)
```

```python
sns.histplot(y_test - y_pred,kde = True )
```
